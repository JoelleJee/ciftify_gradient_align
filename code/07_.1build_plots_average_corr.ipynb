```python
import numpy as np
import os
import pandas as pd
import math
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
analysis_dir = "/Users/Joelle/Anaconda3/envs/BCB430/Gradients/data/analysis/average"
day1 = os.path.join(analysis_dir, "day1_corr.txt")
day2 = os.path.join(analysis_dir, "day2_corr.txt")

```

```python

for d in range(1,3):
    plt.figure()
    
    grad_day = os.path.join(analysis_dir, "day" + str(d) + "_corr.txt")
    grad_day = np.loadtxt(grad_day)
    
    for g in range(6):
        # search for the specific gradient
        grad = grad_day[g, :]

        grad = np.nan_to_num(grad)

        grad = np.arctanh(grad)

        num_subs = grad.shape[0]

        plt.subplot(320 + g + 1)
        sns.kdeplot(grad)
        plt.title("Gradient " + str(g + 1))
        plt.grid(True)

    plt.subplots_adjust(top=0.92, bottom=0.2, 
                        left=0.10, right=0.95, 
                        hspace=0.75, wspace=0.30)

    plt.suptitle("The Density Distribution of \n" +
                 "Correlation Between Average and Subjects Day" + str(d),
                y = 1.15)
    if (d == 1):
        plt.text(-0.07, -1.2, 'Correlation Value', ha='center')
        plt.text(-0.42, 3, 'Number of Subjects', va='center', rotation='vertical')
    else:
        plt.text(-0.06, -1.3, 'Correlation Value', ha='center')
        plt.text(-0.31, 3.1, 'Number of Subjects', va='center', rotation='vertical')
        
    plt.savefig("average_to_subs_day" + str(d) + ".png", bbox_inches='tight')

```
